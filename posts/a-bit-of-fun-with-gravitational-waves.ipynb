# A bit of fun with gravitational waves

```python
import numpy as np
import MotionClouds as mc
fx, fy, ft = mc.get_grids(mc.N_X, mc.N_Y, mc.N_frame)
B_wave = .2
B_v = .2
name = 'waves'
vext = '.webm'
```

```python
def envelope_gravitational(fx, fy, ft, B_wave, g=1.):
    """
     Gravitational envelope:
     selects the plane corresponding to the speed (V_X, V_Y) with some thickness B_V

    """
#    env = np.exp(-.5*((ft**2/g+fx*V_X+fy*V_Y))**2/(B_V*mc.frequency_radius(fx, fy, ft, ft_0=1.))**2)
#    env += np.exp(-.5*((-ft**2/g+fx*V_X+fy*V_Y))**2/(B_V*mc.frequency_radius(fx, fy, ft, ft_0=1.))**2)
    env = np.exp(-.5*((ft**2+np.sqrt(g*(fx**2+fy**2)))**2/(B_v*mc.frequency_radius(fx, fy, ft))**2))
    env += np.exp(-.5*((-ft**2+np.sqrt(g*(fx**2+fy**2)))**2/(B_v*mc.frequency_radius(fx, fy, ft))**2))
#    env = np.exp(-.5*(ft**2-.5**2+(fx -.5)**2)/(B_wave*mc.frequency_radius(fx, fy, ft))**2))
    #env += np.exp(-.5*((-ft**2+(g*(fx**2+fy**2)))/(B_v*mc.frequency_radius(fx, fy, ft))**2))
    return env
```

```python
def envelope_gabor_wave(fx, fy, ft, V_X=mc.V_X, V_Y=mc.V_Y,
                    B_V=mc.B_V, B_wave=B_wave, sf_0=mc.sf_0, B_sf=mc.B_sf, loggabor=mc.loggabor,
                    theta=mc.theta, B_theta=mc.B_theta, alpha=mc.alpha):
    """
    Returns the Motion Cloud kernel

    """
    envelope = mc.envelope_gabor(fx, fy, ft, V_X=V_X, V_Y=V_Y,
                    B_V=B_V, sf_0=sf_0, B_sf=B_sf, loggabor=loggabor,
                    theta=theta, B_theta=B_theta, alpha=alpha)
    envelope += envelope_gravitational(fx, fy, ft, B_wave=B_wave)
    return envelope
```

```python
theta, B_theta, B_v_low, B_v_high = 0., np.pi/3., .5, 1.
loggabor, alpha, sf_0, B_sf, B_V = True, 1., .1, .25, 1.
seed = 12234565
```
