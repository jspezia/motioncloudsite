Movie Contrast Generator
------------------------

```python
def toss():
    return (((np.random.rand()>.5)*2-1))

def mcg(img, nb_test, wait, term, shift_range):
    if (nb_test <= 0):
        return (movie[0, 0, 0])
    period = wait + 2*term
    movie = np.zeros([img.shape[0], img.shape[1], (nb_test*(wait + 2*term))])
    if (img.ndim == 2):
        while (nb_test != 0):
            shift = toss() * np.random.randint(shift_range)
            movie[:, :, ((nb_test-1)*period + wait):((nb_test-1)*period + wait + term)] = img[:, :, np.newaxis]
            imgN = np.roll(img, shift, 1) * toss()
            movie[:, :, ((nb_test-1)*period + wait + term):((nb_test-1)*period + wait + 2*term)] = imgN[:, :, np.newaxis]
            nb_test += -1
    else:
        im1 = img[:, :, ((img.shape[2])/2)]
        while (nb_test != 0):
            shift = toss() * np.random.randint(shift_range)
            movie[:, :, ((nb_test-1)*period + wait):((nb_test-1)*period + wait + term)] = im1[:, :, np.newaxis]
            im2 = img[:, :, ((img.shape[2])/2 + shift)] * toss()
            movie[:, :, ((nb_test-1)*period + wait + term):((nb_test-1)*period + wait + 2*term)] = im2[:, :, np.newaxis]
            nb_test += -1
    return (movie)
```
